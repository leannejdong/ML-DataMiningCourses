```python
import pandas as pd
```

```python
train_data = pd.read_csv('spam_train.csv',skiprows=[0],header=None)
```

```python
train_data.head()
```

```python
detect={'spam':1,'email':0}
```

```python
train_data.iloc[:,58] = [detect[item] for item in train_data.iloc[:,58]]
train_data.head()
```

```python
train_data.to_csv('train_data.csv',header=False, index=False)
```
